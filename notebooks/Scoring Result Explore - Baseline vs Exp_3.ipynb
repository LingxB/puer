## Mistakes baseline made but not exp_3

40 mistakes

```python
def get_pol(w, lx_df):
    try:
        p = round(lx_df.loc[w].mean(),2)
    except KeyError:
        p = None
    return p

def w_in_lx(sent, lx_df):
    plst = [get_pol(w, lx_df) for w in sent.split()]
    cov = sum([1 for p in plst if p is not None])/len(plst)
    return plst, cov

def check_Sn(idx, df, verbose=True, return_cov=False):
    s = df.SENT.iloc[idx]
    pl, cov = w_in_lx(s, lx)
    if verbose:
        print(f'Lx coverage: {cov:.2%}')
    if return_cov:
        return pd.DataFrame([s.split(),pl]), cov
    else:
        return pd.DataFrame([s.split(),pl])
```

```python
tmp = all_mistakes.loc[all_mistakes.PRED_E3.isna()]
exp3_improved = tmp.dropna(axis=1).join(exp3_df.iloc[tmp.index][['PRED','NEG','NEU','POS']], rsuffix='_E3')
exp3_improved
```

```python
len(exp3_improved)
```

```python
check_Sn(0, exp3_improved)
```

```python
check_Sn(1, exp3_improved)
```

```python
check_Sn(2, exp3_improved)
```

```python
check_Sn(3, exp3_improved)
```

As below examples: corrected predictions by adding lexicon, has both average and median of 0.54 lx coverage

```python
np.mean([check_Sn(i, exp3_improved, verbose=True, return_cov=True)[1] for i in range(40)])
```

```python
np.median([check_Sn(i, exp3_improved, verbose=False, return_cov=True)[1] for i in range(40)])
```
